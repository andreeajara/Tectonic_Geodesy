```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
v=0.25
def uz(p,a,E,d,r):
    uz=(2*(1-v**2)*p*a**3*d)/(E*(r**2+d**2)**(3/2))
    return uz
def ur(p,a,E,d,r):
    ur=(2*(1-v**2)*p*a**3*r)/(E*(r**2+d**2)**(3/2))
    return ur
```

```python
p=linspace(10**6,10**7,num=20) 
a=linspace(1000,10000,num=20) #Reinen, L.A. et al. ()
E=linspace(10**10,10**11, num=20) #Stuwe,K. (). Geodynamics of the Lithosphere. pp. 207
d=linspace(2000,40000,num=20)
r=linspace(-30000,30000,num=100)
```

```python
plot(r,uz(p[0],a[10],E[10],d[10],r))
plot(r,uz(p[2],a[10],E[10],d[10],r))
plot(r,uz(p[4],a[10],E[10],d[10],r))
plot(r,uz(p[8],a[10],E[10],d[10],r))
plot(r,uz(p[16],a[10],E[10],d[10],r))
```
