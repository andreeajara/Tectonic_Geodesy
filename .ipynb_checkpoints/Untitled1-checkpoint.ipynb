```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
p=linspace(1,10,num=100) #MPa. Lisowski, 2006
a=linspace(0.5,4,num=100) #km. Pinel, 2003
E=linspace(10**4,10**5, num=100) #MPa. Stuwe,K. (2007). Geodynamics of the Lithosphere. pp. 207
d=linspace(2,30,num=100) #km. Canon-TApia y Szakaks, 2012
r=linspace(-100,100,num=100) #km.
x=linspace(-100,100,num=100) #km.
y=linspace(-100,100,num=100) #km.
theta=23
```

```python
v=0.25 #Constante de Poisson
c=2*(1-v**2) 

#Coordenadas cilindricas. 
def uz(p,a,E,d,r):
    uz=(c*p*a**3*d)/(E*(r**2+d**2)**(3/2))
    return uz

def ur(p,a,E,d,r):
    ur=(c*p*a**3*r)/(E*(r**2+d**2)**(3/2))
    return ur

#Proyeccion Uz en el "look vector".

def Uz(p,a,E,d,r,theta):
    Uz=uz(p,a,E,d,r)*cos(radians(theta))
    return Uz

def UZ(p,a,E,d,x,y):
    uz=(c*p*a**3*d)/(E*(x**2+y**2+d**2)**(3/2))
    return uz

def UR(p,a,E,d,x,y,r):
    ur=(c*p*a**3*r)/(E*(x**2+y**2+d**2)**(3/2))
    return ur
```

```python
X,Y=meshgrid(x,y)
plt.figure(figsize=(10,10))
CS = plt.contour(X, Y, (UZ(p[50],a[50],E[50],d[50],X,Y))*100000)
plt.clabel(CS, inline=1, fontsize=12)
plt.title('Contornos desplazamiento vertical (cm)', fontsize=14)
xlabel('x (km)',fontsize=12)
ylabel('y (km)',fontsize=12)
```

```python
X,Y=meshgrid(x,y)
plt.figure(figsize=(10,10))
CS = plt.contour(X, Y, UR(p[50],a[50],E[50],d[50],X,Y,r)*100000)
plt.clabel(CS, inline=1, fontsize=12)
plt.title('Contornos desplazamiento horizontal (cm)', fontsize=14)
xlabel('x (km)',fontsize=12)
ylabel('y (km)',fontsize=12)
```
